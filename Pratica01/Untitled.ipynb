Transformando em dummie(flag)
--

```python
#transformando as variaveis categoricas em dummies
titanic_encoded = pd.get_dummies(titanic_sem_na, columns=['class','who'],drop_first = True)
titanic_encoded.head(20)
```

Mapping
--

```python
titanic_encoded.sex.unique() 
#mostrando que em 'sex' é definido por male e female
```

```python
titanic_encoded.sex = titanic_encoded.sex.map({'female':1,'male':0})
#passa então que female é 1 e male é 0
```

```python
titanic_encoded.sex.unique()
#mostrando que agora é por 0s e 1s
```

Mudando alguns tipos de dados

```python
titanic_encoded.dtypes
```

```python
titanic_encoded.adult_male.astype(int)
```

```python
titanic_encoded.adult_male = titanic_encoded.adult_male.astype(int)
titanic_encoded.alone = titanic_encoded.alone.astype(int)
#mudando o tipo das colunas para int
```

```python
titanic_encoded.dtypes
```

Seperando as variaveis explicativas da target
--

```python
y = titanic_encoded.survived
```

```python
y
#passando a uma variavel a coluna principal que vamos usar
```

```python
x = titanic_encoded.drop('survived',axis=1)
#outras variaveis explicativas que não vamos usar
```

Árvore de classificação com uma variavel flag
--

```python
clf = DecisionTreeClassifier(random_state=100)
```

```python
x.head()
```

```python
y
```

Homem adulto
--

```python
clf = clf.fit(x[['adult_male']],y)
clf
```

```python
plt.figure(figsize=(25,10))
plot_tree(clf,
         filled=True,
         class_names=['Died','Survived'], #quebra de mortos e vivos
         feature_names=['adult_male']); #baseado no campo homem adulto

#quadro maior, sample = todos os dados
              #value = 461 mortos e 323 vivos

#quadro azul = 333 dados, desses 95 morreram e 238 viveram mulheres
#quadro laranja = 451 dados, desses 366 morreram e 85 viveram homens

#gini é nivel de pureza, quão menor ele for melhor
```

```python
plt.figure(figsize=(25,10))
plot_tree(clf,
         filled=True,
         class_names=['Died','Survived'], #se viveram ou morreram 
         feature_names=['class_Third']); #baseado se eram de 3ª classe
```

Arvore com duas variaveis
--

```python
clf = clf.fit(x[['adult_male','class_Third']],y)
clf
```

```python
plt.figure(figsize=(25, 10))
plot_tree(clf,
          filled=True,
          class_names=['Died', 'Survived'],
          feature_names=['adult_male','class_Third']);

#somando os gini's de toda a arvore mostrada acima, o que for menor é o melhor
#logo, caso for fazer uma arvore de mais de uma variavel, ele pega como a variavel melhor ou seja
#a do topo, a que no seu total tem o gini menor, entre homem adulto e 3 classe, homem adulto tem
#o gini total menor, por isso ele fica em 1 na arvore
```
